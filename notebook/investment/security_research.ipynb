```python
'''
本函数主要是规则化stock的code
从tushare获取stock基本信息存入data\stock_basic_info文件夹内
再从该文件夹读出时，code是int型，如000001这样的code，读出来是1
不符合code规则
'''
def standard_code(x):
    s=str(x)
    if len(s)==2:
        return "0000"+s
    elif len(s)==1:
        return "00000"+s
    elif len(s)==3:
        return "000"+s
    elif len(s)==4:
        return "00"+s
    elif len(s)==5:
        return "0"+s
    elif len(s)==6:
        return s
    else:
        print "code is not standard,code=%r"%(s)
```
