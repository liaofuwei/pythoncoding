# 主要功能：从tushare上下载数据并更新数据
## 1、首先运行##====1=====##，加载各种库
## 2、##====2=====##为获取基本股票信息，并写入listedcompany库中的companybasicinfo表，如果存在companybasicinfo表，则执行更新操作

```python
##====1=====##
from sqlalchemy import create_engine
import tushare as ts
import pandas as pd
```

```python
##====2=====##
engine = create_engine('mysql://root:root@127.0.0.1/listedcompany?charset=utf8')
#先读取数据库中已经存储的companybasicinfo表，用作与此时获取的表进行比较，判断是否需要更新

sql = "select * from companybasicinfo"
tableFromMysql = pd.read_sql(sql,engine).set_index('code')

allCompanyBasicInfo = ts.get_stock_basics() #在线获取的数据
print len(allCompanyBasicInfo),len(tableFromMysql)
if(len(allCompanyBasicInfo)>len(tableFromMysql)):
    allCompanyBasicInfo.to_sql('companybasicinfo',engine,if_exists='append')
    print "update finish!!!!!!!"
else:
    print "mysql database is newest!"

```

```python
allCompanyBasicInfo.index
```

```python
for companycode in allCompanyBasicInfo.index:
    if allCompanyBasicInfo.ix[companycode,'timeToMarket'] == 0:
        df = ts.get_k_data(code = companycode,ktype = 'D',autype = 'qfq',start = '2005-01-01',end = '2016-12-01')
        df.to_sql(companycode,engine)
    else:
        starttime = str(allCompanyBasicInfo.ix[companycode,'timeToMarket'])
        st = starttime[0:4]+"-"+starttime[4:6]+"-"+starttime[6:8]
        df = ts.get_k_data(code = companycode,ktype = 'D',autype = 'qfq',start = st,end = '2016-12-01')
        df.to_sql(companycode,engine)
```
