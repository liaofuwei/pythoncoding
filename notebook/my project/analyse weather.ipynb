```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#批量写入数据库
import MySQLdb
weather_data = pd.read_csv("C:\pythoncoding\pandas\pandas-cookbook\data\weather_2012.csv")
db = MySQLdb.connect("localhost","root","root","test" )
cursor = db.cursor()
sql = "insert into weather2012 values (%s,%s,%s,%s,%s,%s,%s,%s)"
param = []
for i in range(len(weather_data)):
    param.append([weather_data.ix[i,"Date/Time"],weather_data.ix[i,"Temp (C)"],weather_data.ix[i,"Dew Point Temp (C)"],weather_data.ix[i,"Rel Hum (%)"],weather_data.ix[i,"Wind Spd (km/h)"],weather_data.ix[i,"Visibility (km)"],weather_data.ix[i,"Stn Press (kPa)"],weather_data.ix[i,"Weather"]])

cursor.executemany(sql, param)  
db.commit()    
```

```python
weather_data = pd.read_csv("C:\pythoncoding\pandas\pandas-cookbook\data\weather_2012.csv",index_col="Date/Time")
weather_data.index = pd.to_datetime(weather_data.index)
weather_data["date"]=weather_data.index.date
weather_data.head(100)
```

```python
weather_category = weather_data.groupby('Weather').count()
weather_category["sum(hour)"] = weather_category["Temp (C)"]

weather_category = weather_category.ix[:,["sum(hour)"]]
weather_category.plot()
```

```python
weather_data.ix[:,["Temp (C)","Rel Hum (%)"]].plot(figsize=(15, 6),secondary_y = ["Rel Hum (%)"])
plt.show()
```

```python
weather_date = weather_data.groupby('date')["Temp (C)"].mean()
weather_date.plot(figsize=(15, 6))
plt.show()
```
